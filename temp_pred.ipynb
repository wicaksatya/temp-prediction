```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
weather = pd.read_csv('/content/drive/MyDrive/weather1.csv')
weather
```

```python
weather.isnull().any()
```

```python
new_weather = weather.loc[:, [
  'Data.Temperature.Max Temp', 'Data.Temperature.Min Temp',
  'Data.Wind.Direction', 'Data.Temperature.Avg Temp', 'Data.Wind.Speed'
]]
new_weather.head()
```

```python
train_y = new_weather['Data.Temperature.Avg Temp']
train_X = new_weather.loc[:,['Data.Temperature.Max Temp', 'Data.Temperature.Min Temp', 'Data.Wind.Direction', 'Data.Wind.Speed']]
train_X,test_X,train_y,test_y=train_test_split(train_X,train_y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
regression=RandomForestRegressor(max_depth=90,random_state=0,n_estimators=100)
regression.fit(train_X,train_y)
```

```python
prediction=regression.predict(test_X)
np.mean(np.absolute(prediction-test_y))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(test_y, prediction)
mse = mean_squared_error(test_y, prediction)
r2 = r2_score(test_y, prediction)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R2-score: {r2:.2f}")
```

```python
for i in range(len(prediction)):
  prediction[i]=round(prediction[i],2)
pd.DataFrame({'Actual':test_y,'Prediction':prediction,'diff':(test_y-prediction)})
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_absolute_error

# Load your dataset (replace with the actual path)
data_path = '/content/drive/MyDrive/weather1.csv'
# Assuming your dataset has columns 'MaxTemp', 'MinTemp', 'cloudcover', 'humidity', and 'tempC'
# Adjust the column names if needed
data = pd.read_csv(data_path)
target = data['Data.Temperature.Avg Temp']  # Adjust the target column name

# Split data into train and test sets
train_X, test_X, train_y, test_y = train_test_split(
    data[['Data.Temperature.Max Temp', 'Data.Temperature.Min Temp', 'Data.Wind.Direction', 'Data.Wind.Speed']],
    target,
    test_size=0.2,
    random_state=42
)

# Train a Random Forest regressor (your model of interest)
regressor = RandomForestRegressor()
regressor.fit(train_X, train_y)

# Train a Dummy Regressor (baseline)
dummy = DummyRegressor(strategy="mean")
dummy.fit(train_X, train_y)

# Make predictions
predictions_rf = regressor.predict(test_X)
predictions_dummy = dummy.predict(test_X)

# Calculate MAE for both models
mae_rf = mean_absolute_error(test_y, predictions_rf)
mae_dummy = mean_absolute_error(test_y, predictions_dummy)

print(f"Random Forest MAE: {mae_rf:.2f}")
print(f"Dummy Regressor MAE: {mae_dummy:.2f}")
```
